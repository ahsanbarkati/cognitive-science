```python
alpha = 4
mu = 0.001
sigma = 0.01
```

```python
def f(x,y):
    if(x<=0):
        return alpha/(1-x) + y
    if( x > 0 and x < alpha + y):
        return alpha + y
    return -1
```

```python
def gety(y,x):
    return (y - mu*(x+1) + mu*sigma)
```

```python
N=10000
x=np.zeros(N+1)
y=np.zeros(N+1)
for i in range(N):
    x[i+1] = f(x[i],y[i])
    y[i+1] = gety(y[i],x[i])
```

```python
plt.xlabel("n")
plt.ylabel("x_n")
plt.plot(range(N+1),x)
```
